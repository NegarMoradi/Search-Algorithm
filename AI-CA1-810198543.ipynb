## Introduction

#### The purpose of this project is to solve the given problem using search algorithms and determine the differences between uninformed search algorithms (BFS and IDS) and informed search algorithms (A* and weighted A*).

#### At first, we have a table n * m. We may have a potion or a doubling drug in each box. Starting at 0,0, we should go to n-1,m-1.

#### initial state: Each potion is in its own box and the doctor is in box (0,0).                                                                                              The goal is to drink all the potions with minimum cost. 

##### First we define a function to get input. This function builds our table and holds the coordinates of each potion, drug and block. We also have functions to convert a string to a list and a list to a string (this function is used to convert states to sets).
##### We use the success function to determine whether we have reached our goal. 


```python
from time import time
from collections import deque
from math import sqrt
from heapq import heappush
from heapq import heappop


filenames = ['test1.in', 'test2.in', 'test3.in']

def get_input(testname):

    global m
    global n

    f = open(testname, "r")
    nandm = list(f.readline().split(' '))
    n, m = (int(nandm[0]),int(nandm[1]))
    candk = list(f.readline().split(' '))
    c, k = (int(candk[0]),int(candk[1]))

    potions = []
    Ddrug = []
    blocked = []

    for i in range(c):
        potions.append(list(map(int,f.readline().split(' '))))
    
    for i in range(k):
        Ddrug.append(list(map(int,f.readline().split(' '))))

    d = int(f.readline())

    for i in range(d):
        blocked.append(list(map(int, f.readline().split(' '))))

    return potions, Ddrug, blocked

def success(di:list(), pot:list()):
    flag = False
    for i in di:
        if(i[0]!=n-1 or i[1]!=m-1):
            flag = True
            break
    return (flag == False and len(pot) == 0)


def changetos(l):
    s = ''
    if(len(l)==0):
        return s
    for i in l:
        s+=str(i[0])+','+str(i[1])+'-'
    return s[:-1]

def changetol(s:str):
    l = []
    if(len(s)==0):
        return l
    ss = s.split('-')
    r = []
    for i in ss:
        r.append(list(map(int,i.split(','))))
    return r


```

## BFS

##### Doctors, potions, and medicines are stored in a set().  For each doctor, we check all 4 directions (up and down, left and right) and look for potions using BFS. Then we update our variables based on whether the potion or medicine is in each box.
##### To implement the frontier, we use a queue and to explore, we use a set.
##### The par dictionary contains the parent of each state. The all_states variable stores all the states we met
#####  When cost is 1, BFS is optimal.                                                                                                                                                                                     time complexity: O(b^d)                                                                                                                                                                                                           space complexity: O(b^d)  

```python

all_state = 0

def bfs(doc, potions, medicines, blocked:list):
    
    global all_state

    starts = (doc, changetos(potions),changetos(medicines))

    frontier = deque()
    frontier.appendleft(starts)
    explored = set()
    par=dict()

    if(success(doc, potions)):
        return par, starts, starts
    

    while(True):

        if(not frontier):
            return False

        cur = frontier.pop()
        if(cur in explored):
            continue
        explored.add(cur)

        doc,pot,med = cur
        doc,pot,med = changetol(doc),changetol(pot),changetol(med)

        for i in (doc):

            x ,y = i[0],i[1]
        #in pot
            if ([x,y] in pot):
                pot.remove([x,y])
        #in med
            if ([x,y] in med):
                doc.append([n-1,0])
                med.remove([x,y])
        #goal
            if(x == n-1 and y == m-1):
                if(success(doc, pot)):
                    return par, cur, starts
                doc.remove([x,y])
                continue

            #down
            if (x - 1 >= 0):
                if([x-1,y] not in blocked):
                    i[0] = x - 1
                    npot = changetos(pot)
                    nmed = changetos(med)
                    ndoc = changetos(doc)
                    new_state = (ndoc, npot, nmed)
                    i[0] = x
                    if(not new_state in explored):
                        par[new_state]=cur
                        if(success(changetol(ndoc), pot)):
                            return par, new_state, starts
                        frontier.appendleft(new_state)
                        all_state += 1
                    
                   


            #up
            if (x + 1 < n):
                if([x+1,y] not in blocked):
                    i[0] = x + 1
                    npot = changetos(pot)
                    nmed = changetos(med)
                    ndoc = changetos(doc)
                    new_state = (ndoc, npot, nmed)
                    i[0] = x
                    if(not new_state in explored):
                        par[new_state]=cur
                        if(success(changetol(ndoc), pot)):
                            return par, new_state, starts
                        frontier.appendleft(new_state)
                        all_state += 1
                
                

            #left
            if (y - 1 >= 0):
                if([x,y-1] not in blocked):
                    i[1] = y - 1
                    npot = changetos(pot)
                    nmed = changetos(med)
                    ndoc = changetos(doc)
                    new_state = (ndoc, npot, nmed)
                    i[1] = y
                    if(not new_state in explored):
                        par[new_state]=cur
                        if(success(changetol(ndoc), pot)):
                            return par, new_state, starts
                        frontier.appendleft(new_state)
                        all_state += 1
                

            #right
            if (y + 1 < m):
                if([x,y+1] not in blocked):
                    i[1] = y + 1
                    npot = changetos(pot)
                    nmed = changetos(med)
                    ndoc = changetos(doc)
                    new_state = (ndoc, npot, nmed)
                    i[1] = y
                    if( not new_state in explored):
                        par[new_state]=cur
                        if(success(changetol(ndoc), pot)):
                            return par, new_state, starts
                        frontier.appendleft(new_state)
                        all_state += 1
                
```

## IDS

###### In a loop, we give a depth to the DFS function and based on that depth, we move forward until we find the solution. The DFS function takes four arguments: node, nodeparent, depth, and limit and returns one node(final node). A node is a set of states. In this part we use the explored function to store depths.
###### For each doctor, we check all 4 directions (up and down, left and right) and look for potions using DFS. Then we update our variables based on whether the potion or medicine is in each box.  Each series only goes to a depth equal to the size of the limit. If we reach the goal, we return node.
##### To implement the explored, we use a set. The all_states variable stores all the states we met
##### The pathnode dictionary contains the parent of each state.
###### We would return false if we had searched all the depths and reach the limit but still didn't find a solution.
##### Besides not having new_state in explored, we have another condition. We need to check DFS again if the depth of the current state is less than the depth of the new state because we came to this state from a shorter path.



```python
all_states = 0
pathnode = dict() 
explored = dict()
mark = dict()


def dfs(stnode:set(), parent:set(), dep, limit):

    if(dep == limit):
       return False

    global all_states

    all_states = 1 + all_states

    pathnode[stnode] = parent

    if(stnode != parent):
        explored[stnode] = explored[parent] + 1

    
    doc,pot,med = stnode
    doc,pot,med= changetol(doc),changetol(pot),changetol(med)

    mark[stnode] = 1

    if(success(doc, pot)):
        return stnode
    

        

    for i in (doc):

        x ,y = i[0],i[1]
    #in pot
        if ([x,y] in pot):
            pot.remove([x,y])
    #in med
        if ([x,y] in med):
            doc.append([n-1,0])
            med.remove([x,y])
            pathnode[(changetos(doc),changetos(pot),changetos(med))]=parent
    #goal
        if(x == n-1 and y == m-1):
            if(success(doc, pot)):
                return stnode
            doc.remove([x,y])
            continue

        #down
        if (x - 1 >= 0):
            if([x-1,y] not in blocked):
                i[0] = x - 1
                npot = changetos(pot)
                nmed = changetos(med)
                ndoc = changetos(doc)
                new_state = (ndoc, npot, nmed)
                i[0] = x
                if(not new_state in mark or dep+1<explored[new_state]):
                    check = dfs(new_state, stnode, dep+1, limit)
                    if(check):
                        if(not new_state in mark):
                            del mark[stnode]
                        return check
                        


            #up
        if (x + 1 < n):
            if([x+1,y] not in blocked):
                i[0] = x + 1
                npot = changetos(pot)
                nmed = changetos(med)
                ndoc = changetos(doc)
                new_state = (ndoc, npot, nmed)
                i[0] = x
                if(not new_state in mark or dep+1<explored[new_state]):
                    check = dfs(new_state, stnode, dep+1, limit)
                    if(check):
                        if(not new_state in mark):
                            del mark[stnode]
                        return check
                        


            #left
        if (y - 1 >= 0):
            if([x,y-1] not in blocked):
                i[1] = y - 1
                npot = changetos(pot)
                nmed = changetos(med)
                ndoc = changetos(doc)
                new_state = (ndoc, npot, nmed)
                i[1] = y
                if(not new_state in mark or dep+1<explored[new_state]):
                    check = dfs(new_state, stnode, dep+1, limit)
                    if(check):
                        if(not new_state in mark):
                            del mark[stnode]
                        return check
                        


            #right
        if (y + 1 < m):
            if([x,y+1] not in blocked):
                i[1] = y + 1
                npot = changetos(pot)
                nmed = changetos(med)
                ndoc = changetos(doc)
                new_state = (ndoc, npot, nmed)
                i[1] = y
                if(not new_state in mark or dep+1<explored[new_state]):
                    check = dfs(new_state, stnode, dep+1, limit)
                    if(check):
                        if(not new_state in mark):
                            del mark[stnode]
                        return check
        
    return False
                           

def IDS(doc, potions, medicines, blocked:list):
    maxx = 1000
    starts = (doc, changetos(potions),changetos(medicines))
    explored[starts]=0
    for dss in range(100):
        mark.clear()
        pathnode.clear()
        node = dfs(starts,starts,0, dss)
        if(node):
            return node,starts , dss
    return False, False

```

```python

def print_IDS(x,snode):
    
    if(x==snode):
        return
    doc, pot, med= x
    pathp.append(doc)
    print_IDS(pathnode[x],snode)


def printpath_IDS(filename):

    global path
    global pathp
    global blocked

    pathp = []
    t1 = time()
    potions, Ddrug, blocked = get_input(filename)
    doctors = '0,0'
    node, starts, distance = IDS(doctors,potions, Ddrug, blocked)
    if(node!= False):
        print_IDS(node,starts)
    pathp.reverse()
    print(*pathp, sep='\n')
    t2 = time()
    print(t2-t1)
    print('Number of visited states:' + str(all_states))
    print('Distance:' + str(distance-1))
    
print("result for IDS:")

for i in range(len(filenames)):
    print('test' + str(i+1))
    printpath_IDS(filenames[i])
    print("-------------------------------------------------------------------------------------------------------------------")
   
```

## A* 

###### First we define the heuristic() function. The function finds the closest doctor to each potion (d), and returns the potion distance that has the most d. (It finds the potion with the greatest distance from the other doctors.)
##### For a consistent heuristic, the solution is optimal. Our heuristic function is consistent since it is always non-increasing in its path. 

```python
#in A* alpha is 1

def heuristic(doc, pot, alpha):

    minpot = [0]
    docd = []
    for j in pot:
        for i in doc:
            x = sqrt((j[0]-i[0])**2 + (j[1]-i[1])**2)
            docd.append(x)
        minpot.append(min(docd))

    return max(minpot)*alpha
```

###### This function is implemented just like BFS, but the difference is that we always pick the state with the lowest heuristic() value when we pop from the frontier list. In our state variable, we also store the value of heuristic for each node. 
###### For each doctor, we check all 4 directions (up and down, left and right) and look for potions using BFS. Then we update our variables based on whether the potion or medicine is in each box.
###### To implement the frontier, we use a list(heap) and to explore, we use a set.
##### The par dictionary contains the parent of each state. The all_states variable stores all the states we met

```python
           
frontier = []
explored = set()
all_states = int()
par=dict()
mark = dict()

def A_Star(doc, potions, medicines, blocked:list, alpha):
    
    starts = (doc, changetos(potions),changetos(medicines))

    global all_states

    if(success(doc, potions)):
        return starts, starts
    
    heurs = heuristic([[0,0]],potions,alpha)
    #g(n)+h(n)
    heappush(frontier, (0+heurs,starts))
    mark[(0+heurs,starts)] = 1


    while(True):

        if(not frontier):
            return 

        cur1 = heappop(frontier)

        cur, curs = cur1[0],cur1[1]

        if(curs in explored):
            continue

        del mark[cur1]

        doc,pot,med = curs
        doc,pot,med = changetol(doc),changetol(pot),changetol(med)

        if success(doc, pot):
            return curs, starts

        explored.add(curs)

        for i in (doc):

            x ,y = i[0],i[1]
        #in pot
            if ([x,y] in pot):
                pot.remove([x,y])
        #in med
            if ([x,y] in med):
                doc.append([n-1,0])
                med.remove([x,y])
        #goal
            if(x == n-1 and y == m-1):
                if(success(doc, pot)):
                    return  curs, starts
                doc.remove([x,y])
                continue

        #down
            if (x - 1 >= 0):
                if([x-1,y] not in blocked):
                    i[0] = x - 1
                    npot = changetos(pot)
                    nmed = changetos(med)
                    ndoc = changetos(doc)
                    new_state = (ndoc, npot, nmed)
                    i[0] = x

                    if(not new_state in explored):
                        new_hg = cur + 1 + heuristic(changetol(ndoc),pot, alpha) 
                        if(not new_hg in mark):
                            par[new_state]=curs
                            if(success(changetol(ndoc), pot)):
                                return  new_state, starts
                            heappush(frontier, (new_hg,new_state))
                            mark[(new_hg,new_state)] = 1
                            all_states += 1


        #up
            if (x + 1 < n):
                if([x+1,y] not in blocked):
                    i[0] = x + 1
                    npot = changetos(pot)
                    nmed = changetos(med)
                    ndoc = changetos(doc)
                    new_state = (ndoc, npot, nmed)
                    i[0] = x
                    if(not new_state in explored):
                        new_hg = cur + 1 + heuristic(changetol(ndoc),pot, alpha)
                        if(not new_hg in mark):
                            par[new_state]=curs
                            if(success(changetol(ndoc), pot)):
                                return  new_state, starts
                            heappush(frontier, (new_hg,new_state))
                            mark[(new_hg,new_state)] = 1
                            all_states += 1

        #left
            if (y - 1 >= 0):
                if([x,y-1] not in blocked):
                    i[1] = y - 1
                    npot = changetos(pot)
                    nmed = changetos(med)
                    ndoc = changetos(doc)
                    new_state = (ndoc, npot, nmed)
                    i[1] = y
                    if(not new_state in explored):
                        new_hg = cur + 1 + heuristic(changetol(ndoc),pot, alpha) 
                        if(not new_hg in mark):
                            par[new_state]=curs
                            if(success(changetol(ndoc), pot)):
                                return  new_state, starts
                            heappush(frontier, (new_hg,new_state))
                            mark[(new_hg,new_state)] = 1
                            all_states += 1

        #right
            if (y + 1 < m):
                if([x,y+1] not in blocked):
                    i[1] = y + 1
                    npot = changetos(pot)
                    nmed = changetos(med)
                    ndoc = changetos(doc)
                    new_state = (ndoc, npot, nmed)
                    i[1] = y
                    if(not new_state in explored):
                        new_hg = cur + 1 + heuristic(changetol(ndoc),pot, alpha) 
                        if(not new_hg in mark):
                            par[new_state]=curs
                            if(success(changetol(ndoc), pot)):
                                return  new_state, starts
                            heappush(frontier, (new_hg,new_state))
                            mark[(new_hg,new_state)] = 1
                            all_states += 1
```
